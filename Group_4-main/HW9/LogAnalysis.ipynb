# Apache Log Analysis using Python

```python
import pandas as pd
import matplotlib.pyplot as plt
from textwrap import wrap
```

The csv file was obtained using a software called http logs viewer

```python
# filtering the log to only get our entries
df = pd.read_csv("new_logs.csv")
df = df[df['Request'].str.contains("ibenyamna")]
df.shape
```

```python
# to enable us to see the whole table
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df.head()
```

After taking a look at our table, we have some changes to make before we start analyzing the data. <br>
We don't need all the columns since some simply have null values in them, so we have to drop some columns.

```python
df.drop(['Logname','User','Country'], axis=1, inplace=True)
```

```python
# cleaning the date format to make it easier to parse
df['Date'] = df['Date'].apply(lambda x: x.strip("[]"))
df['Date'] = df['Date'].apply(lambda x: x.replace(":"," ",1))
# making the date the index of the dataframe to make time sampling easier to perform
df.index = pd.to_datetime(df.pop('Date'))

df.head()
```

```python
plt.figure(figsize=(10, 8))
df['Code'].value_counts().plot(kind='bar')
plt.title('Number of Occurences of each code')
plt.xlabel("Code")
plt.ylabel("Number of Occurences")

plt.show()
```

```python
plt.figure(figsize=(10, 8))
count = df['UserAgent'].value_counts()
x = count.tolist()
y = count.keys().tolist()

labels = [ '\n'.join(wrap(l, 20)) for l in y ]

plt.figure(figsize=(16, 8))
plt.barh(labels, x)

plt.xlabel("Number of requests")
plt.ylabel("Browser/OS")
plt.title("Number of requests made by different Browsers/OS")
plt.show()
```

```python
count = df['Code'].value_counts()
x = count.tolist()
y = count.keys().tolist()

plt.figure(figsize=(16, 8))
patches, texts = plt.pie(x, labels=y)
plt.title('Percentage of Occurences of each code')

total = sum(x)
percentages = list(map(lambda val: 100*val/total, x))

labels = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(y, percentages)]

plt.legend(patches, labels, fontsize=10, prop={'size':12})
plt.show()
```

```python
# plotting the bytes of transfered data over the whole log
plt.figure(figsize=(16, 8))
plt.title("Bytes transfered over the whole log in 3 hours intervals")
df_s = df['Size'].resample('3h').sum()
df_s.plot()
plt.ylabel("Bytes")
plt.show()
```

```python
# sampling the number of requests over 5 minutes intervals during a day ('11/11/2021') 
plt.figure(figsize=(16, 8))
plt.title('Number of requests over 5 minutes intervals during 11/11/2021')
df_s = df['Request'].resample('5T').count()
plt.ylabel('Requests')
df_s['11/11/2021'].plot()
plt.show()
```

```python
# sampling the number of requests over a month using day intervals, November
plt.figure(figsize=(16, 8))
plt.title('Number of requests over 11/2021 using one day intervals')
df_s = df['Request'].resample('D').count()
plt.ylabel('Requests')
df_s['11/2021'].plot()
plt.show()
```

```python
# number of unique users during one month
plt.figure(figsize=(16, 8))
df_s = df['IP Address'].resample('D').nunique()
plt.ylabel('# of unique IP Addresses')
df_s['11/2021'].plot()
plt.title('Number of unique users during one month')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.title('Number of occurences of each unique IP Address')
plt.ylabel('IP Address')
plt.xlabel('# of occurences')
df['IP Address'].value_counts().plot.barh()
plt.show()
```
